```python
import requests
```

```python
import bs4
```

```python
url = 'https://lt.wikipedia.org/wiki/S%C4%85ra%C5%A1as:Lietuvos_%C5%BEuvys'
```

```python
response = requests.get(url)
```

```python
response.text
```

```python
soup = bs4.BeautifulSoup(response.text)
```

```python
flist = soup.find("ol")
```

```python
flist = flist.find_all("li")
```

```python
flist[0].find("a").text
```

```python
zuvys = []
for zuvis in flist:
    tekstas = zuvis.find("a").text
    zuvys.append(tekstas)
```

```python
zuvys
```

```python
import pandas
```

```python
data = pandas.DataFrame(zuvys, columns=['zuvis'])
```

```python
data.to_excel("zuvys.xlsx")
data.to_csv("zuvys.csv")
```
